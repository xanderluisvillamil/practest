```python
def calculate_average(grades_list):
  if not grades_list:
    return 0
  total_sum = sum(grades_list)
  average = total_sum / len(grades_list)
  return average

name  = input("What is your name?")
print("Hello", name)

all_grades = []
num_subjects = int(input("How many subjects will you input?"))

for i in range(num_subjects):
  grade = int(input(f"Enter your grade for subject {i+1}: "))
  all_grades.append(grade)

print("Your total average is", calculate_average(all_grades))
```
